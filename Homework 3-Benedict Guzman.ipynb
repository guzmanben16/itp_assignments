```python
def missing_values_col(): 
    NHL_Goalies=pd.read_excel('NHLGoalies2016_2017.xls',na_values='',sheetname=0) 
    NHL_Goalies=NHL_Goalies.replace('NaN',np.nan) #Make sure that NaN Null must be specified using numpy
    NHL_Series=NHL_Goalies.isnull().sum() #Add all NA's from each column. In this case, all of the values here are in a series and must be converted into a dataframe. 
    return NHL_Series.to_frame(name="Missing") # Convert the series into a dataframe

missing_values_col()
```
